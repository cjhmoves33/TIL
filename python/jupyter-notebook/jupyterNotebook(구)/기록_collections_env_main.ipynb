```python
#################### collections ####################

import collections

#collections.deque()

#deque(데쿠,데크)는 큐(Queue)구조를 사용할 수 있게 만든 리스트라고 보면 된다.
#리스트에 비해 효율적인 자료저장을 지원한다.
deq_1 = collections.deque()
deq_1.append("first")
deq_1.append(1)
>> deque(['first', 1])
deq_1.insert(0,"zero")
deq_1.extend([2,3,4,"five"])
>> deque(['zero', 'first', 1, 2, 3, 4, 'five'])
#작동 방식이 리스트와 흡사하다.
#또, deque는 Linked List의 특성을 지원한다.
#자세히는 singly linked list, doubly linked list, Circular Doubly Linked List이렇게 세가지로 나뉘는데,
#파이썬의 list는 기본적으로 linked list이고 데크는 Circular Doubly Linked List 방식을 지원하게 만들어진거 같다..
#reverse와 rotate를 할 수있기 때문이다.

a = collections.deque({"name":"jaeha", "age":25}.items())
deq_1.extendleft(a)
deq_1.insert(0, "onetwo")
deq_1.insert(0, ["one","two"])
>> deque([['one', 'two'], 'onetwo', ('age', 25), ('name', 'jaeha'), 'zero', 'first', 1, 2, 3, 4, 'five'])
#extend,extendleft에는 iterable이기만 하면 값으로 넣을 수 있는데, 딕셔너리.items()는 튜플로된 리스트를 반환하기에 가능하다.
#위 코드를 봤을 때 extendleft의 경우에는 인자를 iterable로 받게되고, insert는 값x를 받는데, 
#extendleft로 들어온 값의 순서가 반전되는걸 볼 수있다.
#인자값으로 iterable을 받는것과 값이 반전된걸 보면 코드의 효율성을 위한 설계가 된것임을 추측할 수 있다.(iterabl의 0번째 값을 데크의 0번째로 추가)
deq_1.reverse()
>> deque(['five', 4, 3, 2, 1, 'first', 'zero', ('name', 'jaeha'), ('age', 25), 'onetwo', ['one', 'two']])
deq_1.rotate(2)
>> deque([4, 'five', ['one', 'two'], 'onetwo', ('age', 25), ('name', 'jaeha'), 'zero', 'first', 1, 2, 3])
deq_1.rotate(-3)
>> deque([('name', 'jaeha'), 'zero', 'first', 1, 2, 3, 4, 'five', ['one', 'two'], 'onetwo', ('age', 25)])


#collections.OrderdDict
#파이썬 3.6부터는 얘기가 달라지지만, 기존의(3.6전)파이썬에서는 딕셔너리가 값의 순서를 보장하지못했다.
#그래서 OrderdDict를 사용해서 순서대로 딕셔너리값을 저장했지만, 3.6부터는 딕셔너리가 값의 순서를 보장하기 때문에 사용할 필요는없다.

```

```python


#collections.defaultdict()
#디폴트딕셔너리는 키를 입력했을 때 값을 따로입력하지않아도 자동으로 지정하는게 가능하다.

dict_z = collections.defaultdict(lambda : 0)
dict_i = collections.defaultdict(int)
dict_l = collections.defaultdict(list)
dict_s = collections.defaultdict(set)
#디폴트딕트를 생성할 때 인자값으로 리턴값을 가지는 함수(매개변수x)를 사용해도되고, 자료형을 넣어줘도 된다.
#dict_z 와 dict_i 는 같은 값을 가지게 된다.



s = "hello, im ironman.thanks"
for i in s:
    dict_i[i] += 1
s = sorted(dict_i.items(), key=lambda x: x[1], reverse=True)

l = [('yellow', 1), ('blue', 2), ('yellow', 3), ('blue', 4), ('red', 1)]
for k, v in l:
    dict_l[k].append(v)

d = [{"nm":"bakk", "kda":3,}, {"nm":"bakk", "kda":3,}, {"nm":"aksi", "kda":7}, {"nm":"jlt", "kda":4}]
for i in d:
    for k, v in i.items():
        dict_s[k].add(v)
print(dict_l)
print(s)
print(dict_s)
>> [('n', 3), ('h', 2), ('l', 2), ('o', 2), (' ', 2), ('i', 2), ('m', 2), ('a', 2), ('e', 1), (',', 1), ('r', 1), ('.', 1), ('t', 1), ('k', 1), ('s', 1)]
defaultdict(<class 'list'>, {'yellow': [1, 3], 'blue': [2, 4], 'red': [1]})
defaultdict(<class 'set'>, {'nm': {'bakk', 'aksi', 'jlt'}, 'kda': {3, 4, 7}})
#조금 의문스러운 부분이 있었는데, 두번째와 세번째 코드에서 딕트의 key부분이 지정이 안됐을텐데 append와 add를 사용했다는거다.
#python docs의 설명을 읽어보면, 디폴트딕트는 key가 발견이 되기만하면 바로 딕셔너리를 생성하는거같다.

```

```python


#collections.Counter()
#인자를 iterable로 받고 그 안의 element들이 hashable이면 갯수를 세준다.

from collections import Counter
li_hable = [('one'),"a","b","c","d","a"]
C = Counter(li_hable)
li_nhable = [['one'],"a","b","c","d","a"]
print(Counter(li_hable))
print(C["Two"])
print(Counter(li_nhable))
>> Counter({'a': 2, 'one': 1, 'b': 1, 'c': 1, 'd': 1})
   0
   TypeError: unhashable type: 'list'

#첫번째는 엘리멘트가 전부 hashable이고, 두번째는 hashable이아닌 list가 포함돼있다(에러).
#Counter는 존재하지않는 항목을 키인덱싱을 할때 0을 반환한다는 점을 제외하면 딕셔너리와 같은 인터페이스를 같는다.


from collections import Counter
from pathlib import Path
import csv
import re
p = Path("../EXs/md_csv_example_w.csv")
r = csv.reader(p.open("r",encoding="utf8"), delimiter=",", quotechar='"')
word_li = []
for i in r:
    word = re.findall(r"\w+",str(i))
    for ii in word:
        word_li.append(ii)
Counter(word_li).most_common(5)
>> [('0', 58), ('FL', 10), ('COLUMBIA', 10), ('COUNTY', 10), ('30', 10)]
#csv파일을 불러온뒤 각 필드(줄)을 re 모듈로 필요한값들만(단어) 빼냈다.
#re.findall()에 들어간인자는 문자열리스트로 반환되는데, 그걸 리스트(word_li)에 순서대로 넣게되면 리스트를 요소로갖는
#이차원리스트가 된다. 요소가 리스트(no hashable)면 Counter로 숫자를 셀 수가 없기때문에,
#re.findall()로 반환된값들을 다시 for문으로 빼내어서 word_list로 넣어줬다.
#Counter().most_common(n)은 쉽게생각해서 개수가 가장많은 요소를 top n 까지 보여주겠다는거다.
#물론 반환값은 Counter와는 다르게 튜플로 이루어진 리스트 이다.

#참고로 집합연산도 가능하다.
```
