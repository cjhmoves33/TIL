```python
class NoteBook():
    """Urth.CO -- New Generation Note Book"""

    __company = "Urth"
    __how_many_making = 0
    __os_version = f"OS_version : {1.0}"

    def __init__(self, name, size, color="black", product_ver=0):
        self.__name = name
        self.__product_ver = product_ver
        self.__size = size
        self.__color = color
        NoteBook.__how_many_making += 1

    @classmethod
    def os(cls):
        """ get os_version """
        print(cls.__os_version)
    @classmethod
    def os_set(cls, setting:float):
        """version update"""
        cls.__os_version = f"OS_version : {setting}"
    @classmethod
    def many(cls):
        print(cls.__how_many_making)
    @property
    def information(self):
        self.__info = {
            "COMPANY":NoteBook.__company, "PRODUCT NAME":self.__name, "SIZE":self.__size,
             "PRODUCT_VERSION":self.__product_ver, "COLOR":self.__color
            }
        return self.__info
#클래스의 데코레이터를 보면 @classmethod라는 것들이 보인다. 이는 내장 데코레이터중 하나로 클래스를 self인자로 받는 데코레이터다.
#클래스의 메소드들이 자동으로 인스턴스를 첫번째인자로 받을 때 관례상 self를 쓰는 것과 마찬가지로 classmethod는 cls(class는 예약어라 쓸 수 없기에)를 쓴다.
#클래스의 특성상 반드시 classmethod를 써야만 위 코드의 기능을 수행할 수 있는것은아니지만,
#조금더 직관적으로 사용할 수 있고, 사용자는 단순코드만으로 던더변수,함수(private변수,함수)에 접근할 수 있다.
my_notebook = NoteBook("NoteBook_Class_A", 2)
my_notebook.os()
>>  OS_version : 1.0
#인스턴스에서 클래스메소드를 실행했는데도 값이 출력된다. 위에서 설명 했던거와 같이 인스턴스는 자신이 가지지않은 정도를 클래스에서 가져올 수 있다.
#인스턴스에 클래스가 NoteBook이란 정보를 가지고 있고 클래스를 인자로 준것.


#다음으로 보이는 @property라는 데코레이터다 이 또한 내장 데코레이터중 하나로 property()와 같은 역할을 수행한다.
#@property는 함수를 atrribute처럼 사용할 수 있게 해준다.
my_notebook.information
>>  {'COMPANY': 'Urth',
 'PRODUCT NAME': 'NoteBook_Class_A',
 'SIZE': 2,
 'PRODUCT_VERSION': 0,
 'COLOR': 'black'}
#함수를 attribute처럼 사용했다. 심지어 던더변수를 불러들인것이다. 사용자의 측면에서는 attribute를 사용한것과 같은 느낌을 받는 것이다.
```

```python
class Urth1(NoteBook):
    __number = 0
    
    def __init__(self, name, size, color, product_ver):
        super().__init__(name, size, color, product_ver)
        Urth1.__number += 1
        self.__number = Urth1.__number
        self._user_name = None

    def __greeting(self):
        print(f"hi,{self._user_name} 👋👋")

    @property
    def user(self):
        print(f"USER : '{self._user_name}'")
    @user.setter
    def user(self, new_name):
        # print(f"before - {self._user_name}")
        self._user_name = new_name
        # print(f"after - {self._user_name}")
    @property
    def information(self):
        inform = super().information
        inform["NUMBER"] = self.__number
        inform["USER"] = self._user_name
        return inform
#NoteBook을 상속받은 자식클래스 Urth1을 만들었다. __init__내부를 보면 Urth1.__nunmber가 보인다. 클래스변수인 __number를 사용하기 위한 것이다.
#부모클래스에서 상속받은 __init__함수에 새로운 기능을 추가하고 싶어져서 다시 __init__을 설정해줬다. 이렇게 자식클래스에서 똑같은 이름의 함수지만 
#다른 기능을 구현하기위해 다시 쓰여진것을 오버라이딩(overriding)이라고 한다.
#오버라이딩을 했지만 부모클래스의 초기화함수는 그대로 가지고오기위해 super()를 사용해준다.
#super()는 부모클래스에 접속할 수 있는 프록시를 돌려준다. 부모클래스와 접촉할 수 있는 인터페이스인 셈이다.


my_notebook2 = Urth1("NoteBook_Class_E", 24, "green", 1.2)
my_notebook2._Urth1__greeting()
>>  hi,None 👋👋
#아직 _user_name은 None인 상태, _Urth1__greeting()를 사용해서 숨겨진(?) 함수를 실행시켜줬다.

#Urth1클래스에서도 @property를 사용했는데 추가된 것이 하나 있다. 바로 '@user.setter'
#propety는 기본적으로 getter와 setter를 가지는데 이를 데코레이터로 사용하게 되면
#@propery로 사용된 코드가 getter, @getter_func.setter가 setter가 된다.
my_notebook2.user = "Master Choi"
my_notebook2.user
>>  USER : 'Master Choi'
#상당히 아름다운 기능이다. getter, setter외에도 deleter, doc 의 기능도 사용가능하다.

my_notebook2.information
>>  {'COMPANY': 'Urth',
 'PRODUCT NAME': 'NoteBook_Class_E',
 'SIZE': 24,
 'PRODUCT_VERSION': 1.2,
 'COLOR': 'green',
 'NUMBER': 1,
 'USER': 'Master Choi'}
#imformation또한 오버라이딩을 한경우인데, 부모클래스의 함수를 super.information으로 받아온뒤, 객체화했다.(딕셔너리)
#super()로 인해 부모클래스에 있는 information을 그대로 사용한뒤 객체화 한것이다.
```
