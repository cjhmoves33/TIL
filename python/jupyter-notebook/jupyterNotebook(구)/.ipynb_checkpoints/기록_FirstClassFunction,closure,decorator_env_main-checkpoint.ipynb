```python
#################### first class function, closure ####################

#first class function
#일급함수(일등합수)는 함수를 함수의인자로 전달할 수 있고, 함수의 리턴값으로 함수자체를 반환할 수 있다.
#또한, 함수를 변수에 할당하거나 데이터 구조안에 저장 할 수 있다.

def plus(a ,b):
    print( a + b)
pl = plus
plus(1,2)
print(plus)
print(pl)
pl(1,2)
>>  3
    <function plus at 0x7f8050085f70>
    <function plus at 0x7f8050085f70>
    3
#plus라는 함수를 변수에다 대입을 했을 때 주소값이 같은것을 확인할 수 있다.
#물론 실제 함수를 실행하는 것처럼 사용 했을 때의 값도 실제함수를 실행한것과 같이 나온다.

def plus(a ,b):
    return a + b

def map_plus(method, arr):
    result = [method(i, i) for i in arr]
    print(result)
    return result
li = [1,2,3,4,5]
map_plus(plus, li)
>> [2, 4, 6, 8, 10]
#map_plus 함수는 인자로 plus함수와 리스트를 받았다. 이렇게 간단한 코드에서는 굳이 이렇게 까지 할 필요가 있나싶지만
#다양한 함수들을(결과를 정상적으로 낼 수 있는) map_plus함수에 인자로 줘서 같은규격으로 다양한 값을 낼 수 있다는 장점이있다.

def logger(msg: str):
    def log_msg():
        print("DEBUG : ", msg)
    return log_msg
deb_log = logger("starting")
print(deb_log)
deb_log()
>>  <function logger.<locals>.log_msg at 0x7f93c00525e0>
    DEBUG :  starting
#함수안의 log_msg함수를 리턴해주는 함수를 변수에 할당하고 그 변수를 이용해서 함수를 호출했다.
#여기서 특이한 점은 msg에 str인자를 넣어줘서 함수내 지역변수로서 작용하던것은 함수가 한번 호출되고나면 메모리상에서 지워지고 다시 참조를 할 수 없는데,
#deb_log()로 다시 호출했더니 starting을 기억하고있는게 보인다.
#log_msg같은 함수를 클로저(closure)라고 부른다.
#클로저는 다른함수의 지역변수를 함수가 종료된 후에도 저장한다.

del logger
try:
    print(logger)
except NameError as ne:
    deb_log()
    print(ne)
>>  DEBUG :  starting
    name 'logger' is not defined
#del로 logger자체를 지웠는데도 "starting"을 기억하고 있다.

def tagging(tag : str):
    def txt(content : str):
        print(f"<{tag}>{content}<{tag}>")
    return txt
tag_h1 = tagging("h1")
tag_div = tagging("div")
tag_h1("im ironMan")
tag_div("be flexible area")
>>  <h1>im ironMan<h1>
    <div>be flexible area<div>
#지역변수를 기억하는 클로저함수를 사용해서 같은 양식의 다른 객체들을 생성했다.

```

```python

#closure

#프로그래밍 언어에서의 클로저란 퍼스트클래스 함수를 지원하는 언어의 네임 바인딩 기술이다.
#클로저는 어떤 함수를 함수 자신이 가지고 있는 환경과 함께 저장한 레코드이다.
#또한 함수가 가진 프리변수(free variable)를 클로저가 만들어지는 당시의 값과 레퍼런스에 맵핑하여 주는 역할을 한다.
#클로저는 일반 함수와는 다르게, 자신의 영역 밖에서 호출된 함수의 변수값과 레퍼런스를 복사하고 저장한 뒤, 이 캡처한 값들에 액세스할 수 있게 도와준다.

def freevar(low):
    
    header = "Same"
    def upp(up):
        result = [{header: word}for word in zip(low, up)]
        print(result)
    return upp
lw = ["life","is","short"]
upp = ["LIFE","IS","SHORT"]
samesame = freevar(lw)
samesame(upp)
>>  [{'Same': ('life', 'LIFE')}, {'Same': ('is', 'IS')}, {'Same': ('short', 'SHORT')}]
#프리변수는 상대적인 개념인데 클로저인 upp의 입장에서는 low로 들어올는 값, header는 프리변수가된다.
#upp의 기준에서는 자신의 블록내에서 정의된것도 아니고, 글로벌변수도 아니지만 freevar라는 함수안에서 정의된 변수이기 때문에 upp기준에서는
#프리변수 인것이고, freevar입장에서는 지역변수 인것이다.
#위 파트에서도 봤듯이 upp은 클로저이고 samesame함수는 클로저로 만들어졌다. 이미 samesame = freevar(lw) 를 선언했었기에
#클로저 자체적으로 선언됐던 변수값, 레퍼런스들을 자체적으로 전부 기억한다.(심지어 freevar를 del해도 기억한다.)

```

```python


#decorator

#데코레이터는 클로저와 유사하지만 다른함수의 인자로 함수를 전달한다는 점에서 차이점이 있다.
def deco(display):
    def wrapper(*args, **kwargs):
        print(f"function'{display.__name__}' not call yet")
        return display(*args, **kwargs)
    return wrapper
@deco
def info_display(name, age):
    print(f"{__name__} called successfully")
    print(f"{name}, {age}")

info_display("kevin", 20)
>>  function'info_display' not call yet
    __main__ called successfully
    kevin, 20
#데코레이터는 클로저를 만드는 과정이 @데코레이터함수명 을 추가하면서 클로저를만드는 과정을 생략한 것이다.
#데코레이터가 만들어지는 과정이 헷갈리면 클로저를 만든다고 생각한 후, 클로저를 선언해주는 과정만 빼면된다.
#위 코드 같은경우,
info_display = deco(info_diplay) 
#클로저생성, 사실상 info_display == wrapper
info_display("kevin", 20)
#클로저 생성과정만 사라진 것이다.

#좀더 직관적으로 표현하자면
def deco(info_display):
    def wrapper(name, age):
        print(f"function'{info_display.__name__}' not call yet")
        return info_display(name, age)
    return wrapper
@deco
def info_display(name, age):
    print(f"{__name__} called successfully")
    print(f"{name}, {age}")

info_display("kevin", 20)
#이런 데코레이터로 인해 이런 형식이 되는 것이다.

import logging
def log_deco(STREAM_info):
    FORMAT = "%(asctime)s %(name)s %(levelname)s %(message)s"
    logging.basicConfig(level=logging.DEBUG)
    formatter = logging.Formatter(fmt=FORMAT)
    fhld = logging.FileHandler(filename="tester.log", mode="a", encoding="utf8")
    fhld.setFormatter(formatter)
    logger = logging.getLogger("lgg")
    logger.addHandler(fhld)        
    def wrapper(*args, **kwargs):
        logger.info(f"{STREAM_info.__name__} starting")
        return STREAM_info(*args, **kwargs)
    return wrapper
@log_deco
def info_streamer(user, name):
    inf_st = logging.getLogger("information")
    inf_st.addHandler(logging.StreamHandler())
    inf_st.setLevel(level=logging.INFO)
    inf_st.info(f"{name} get nickname ->{user}")
    
info_streamer(name="TONY", user="ironman")
#로깅함수에도 데코레이터가 잘작동한다.


import logging
import time
import datetime

def log_deco(info):
    logging.basicConfig(
        level=logging.DEBUG, filename=f"{info.__name__}.log", filemode="a",
        format="%(asctime)s %(levelname)s %(message)s" ,datefmt= "%H:%M:%S"
        )
    logger = logging.getLogger("$$ debugger $$")

    def wrapper(*args, **kwargs):
        logger.debug(f"'{info.__name__}' starting")
        return info(*args, **kwargs)
    return wrapper

def log_timer(info):
    t1 = datetime.datetime.now().strftime("%H:%M:%S")
    def timer(*args, **kwargs):
        t2 = time.time()
        time.sleep(0.05)
        t3 = time.time()-t2
        print(f"'{info.__name__}'함수 실행에 걸리는 시간: {t3:.10f} 초", f"함수 생성시간- {t1}")
        return info(*args, **kwargs)
    return timer

@log_deco
def info_streamer(user, name):
    inf_st = logging.getLogger("$$ information $$")
    inf_st.info(f"{name} get nickname -> '{user}'")    

info_streamer(name="TONY", user="ironman")


@log_timer
@log_deco
def time_tester(name, age):
    print(f"display {name}--{age} working!")

time_tester("stark", 40)
#데코레이터의 기본설정상 데코레이터2개를 적용시키게 되면 아래의 데코레이터부터(여기선 @log_deco)적용하고 그 적용한 함수(여기선 wrapper)를 
#위의 데코레이터 (여기선 log_timer)에 적용하게 된다!

>>  'time_tester'함수 실행에 걸리는 시간: 0.0538990498 초 함수 생성시간- 10:18:47
    display stark--40 working!
    
    'wrapper'함수 실행에 걸리는 시간: 0.0544631481 초 함수 생성시간- 10:19:16
    display stark--40 working!
#이 결과는 time_tester함수에 @log_timer만 적용했을 때와 위 코드와같이 두개의 데코레이터를 적용했을 때의 차이점을 명확히 보여준다.
```

```python


import logging
import time
import datetime

def log_deco(info):
    logging.basicConfig(
        level=logging.DEBUG, filename=f"{info.__name__}.log", filemode="a",
        format="%(asctime)s %(levelname)s %(message)s" ,datefmt= "%H:%M:%S"
        )
    logger = logging.getLogger("$$ debugger $$")

    def wrapper(*args, **kwargs):
        logger.debug(f"'{info.__name__}' starting")
        return info(*args, **kwargs)
    return wrapper

def log_timer(info):
    t1 = datetime.datetime.now().strftime("%H:%M:%S")
    def timer(*args, **kwargs):
        t2 = time.time()
        time.sleep(0.05)
        t3 = time.time()-t2
        print(f"'{info.__name__}'함수 실행에 걸리는 시간: {t3:.10f} 초", f"함수 생성시간- {t1}")
        return info(*args, **kwargs)
    return timer

@log_timer
@log_deco
def time_tester(name, age):
    """[tester]
    Args:
        name ([str]): [user name]
        age ([int]): [user age]]
    """
    print(f"display {name}--{age} working!")

time_tester("stark", 40)
print(time_tester.__name__)
print(time_tester.__doc__)
>>  'wrapper'함수 실행에 걸리는 시간: 0.0550618172 초 함수 생성시간- 10:44:33
    display stark--40 working!
    timer
    None
#함수가 데코레이터를 돌게 되면 자신의 이름과 doc을 잃어버리는게 보인다.
#이를 방지 할 수 있는데 파이썬 내장모듈인 functools.wraps을 사용하면 된다.



import logging
import time
import datetime
from functools import wraps

def log_deco(info):
    logging.basicConfig(
        level=logging.DEBUG, filename=f"{info.__name__}.log", filemode="a",
        format="%(asctime)s %(levelname)s %(message)s" ,datefmt= "%H:%M:%S"
        )
    logger = logging.getLogger("$$ debugger $$")

    @wraps(info)
    def wrapper(*args, **kwargs):
        logger.debug(f"'{info.__name__}' starting")
        return info(*args, **kwargs)
    return wrapper

def log_timer(info):
    t1 = datetime.datetime.now().strftime("%H:%M:%S")
    @wraps(info)
    def timer(*args, **kwargs):
        t2 = time.time()
        time.sleep(0.05)
        t3 = time.time()-t2
        print(f"'{info.__name__}'함수 실행에 걸리는 시간: {t3:.10f} 초", f"함수 생성시간- {t1}")
        return info(*args, **kwargs)
    return timer

@log_timer
@log_deco
def time_tester(name, age):
    """[tester]
    Args:
        name ([str]): [user name]
        age ([int]): [user age]
    """
    print(f"display {name}--{age} working!")

time_tester("stark", 40)
print(time_tester.__name__)
print(time_tester.__doc__)
>>  'time_tester'함수 실행에 걸리는 시간: 0.0544569492 초 함수 생성시간- 11:31:19
    display stark--40 working!
    time_tester
    [tester]
        Args:
            name ([str]): [user name]
            age ([int]): [user age]
#@wraps를 사용해서 wrapped함수의 특정데이터들을 잃지않을 수 있다.
#https://docs.python.org/ko/3/library/functools.html#functools.update_wrapper
```
