```python
#################### sorted() (built-in) ####################

#리스트만 정렬를 할 수 있는 sort()와는 다르게 sorted는 iterable을 정렬한 후 리스트로 반환해준다.
#list.sort()는 리턴없이 list를 변경하고, sorted(iterable)은 iterable을 변경하지않고 리턴값이있다.
a = [1,4,5,3,7]
b = sorted(a)
print(a)
print(b)
>> [1, 4, 5, 3, 7]
   [1, 3, 4, 5, 7]



score = [
    {"name":"kevin", "math":60, "ko":70, "en":80},
    {"name":"brian", "math":50, "ko":80, "en":30},
    {"name":"nicolas", "math":90, "ko":70, "en":90}
]

avr_score = list()
for i in score:
    name = i["name"]
    score = (i["math"], i["ko"], i["en"])
    per_score = f"{sum(score)/3:.2f}"
    n_s = tuple([name, float(per_score)])
    avr_score.append(n_s)

rating = sorted(avr_score, key=lambda score:score[1], reverse=True)
print(rating) 
>> [('nicolas', 83.33), ('kevin', 70.0), ('brian', 53.33)]
#먼저 딕셔너리가 든 리스트에서 학생별 점수의 평균을 구한 다음, 또 다른 리스트에 튜플형식("이름",점수)으로 저정했다.
#솔팅을 하기위해 위에서 만들 이터러블 객체인 avr_score를 입력
#다음으로 솔팅을 할 기준이 될 key에는 "단일 매개변수를 인자값으로 취하고 정렬을위해 사용될 값을 반환하는 함수" 또는 callable을 입력해준다.
#위에서 키를 정한것 처럼 일반적으로 인덱스를 키로 쓴다.
#reverse는 오름차순, 내림차순으로 할지 정할 수 있다.
```

```python




from operator import itemgetter, attrgetter

score = [
    {"name":"kevin", "math":60, "ko":70, "en":80},
    {"name":"brian", "math":50, "ko":80, "en":30},
    {"name":"nicolas", "math":90, "ko":70, "en":90}
]

avr_score = list()
for i in score:
    name = i["name"]
    score = (i["math"], i["ko"], i["en"])
    per_score = f"{sum(score)/3:.2f}"
    n_s = tuple([name, float(per_score)])
    avr_score.append(n_s)


class Student:
    def __init__(self, name, grade, age) -> None:
        self.name = name
        self.grade = grade
        self.age = age
    
    def __repr__(self) -> str:
        return repr((self.name, self.grade, self.age))

mrk = Student("K", "A", 15)
mrG = Student("G", "B", 15)
mrH = Student("H", "A", 14)
mrI = Student("I", "C", 14)

s_list = [mrk, mrG, mrH, mrI]


rating_sort = sorted(avr_score, key=itemgetter(1), reverse=True)
age_sort = sorted(s_list, key=attrgetter("age"), reverse= True)

print(rating_sort)
print(age_sort)

>> [('nicolas', 83.33), ('kevin', 70.0), ('brian', 53.33)]
   [('K', 'A', 15), ('G', 'B', 15), ('H', 'A', 14), ('I', 'C', 14)]
    
#여기서 sorted함수를 사용한 부분을 보면 key에 operator모듈에서 불러온 함수를 사용했다.
# itemgetter(item or *items) - "피연산자의 __getitem__() 메서드를 사용하여 피연산자에서 item을 꺼내는 콜러블 객체를 반환합니다." 여러 항목이 지정되면, 조회 값의 튜플을 반환합니다.
# attrgetter(attr or *attrs) - "피연산자에서 attr을 꺼내는 콜러블 객체를 반환합니다." 둘 이상의 어트리뷰트가 요청되면, 어트리뷰트의 튜플을 반환합니다. 어트리뷰트 이름은 점을 포함할 수도 있습니다
#둘다 인자값과 상호작용이 가능한 콜러블 객체를 반환하는 함수이다.
#위에서 설명했듯이.key에는 콜러블을 넣어줘도 무관하다.

#f = attrgetter('name') 다음에, f(b) 호출은 b.name을 반환합니다.
#f = itemgetter(2) 다음에, f(r) 호출은 r[2]를 반환합니다.
#⬆️작동 방식
```

```python
#callable에 대해서 찾아보던 중에 그럴싸한 생각이났다.
#key = lamda x: x[1], key=itemgetter(1) 둘다 함수를 객체로 옮겨주는 방법과 동일하다.
#사실 람다함수를 객체로 옮기든 일반함수를 객체로 옮기던 callable임에는 틀림이 없다.
#또, 바로위 작동방식을 보면 key에 함수(또는 다른 콜러블)를 넣는방식과 동일하다.

결론(추측..) - sorted 내장함수를 차근차근 살펴보면
sorted(iterable, **kwargs) 로 구성이됐다는 것을 짐작할 수 있다.
안을 들여다보면 분명 "key"와 "reverse"라는 인자를 필요로하는 부분이 있을 것이다.
그리고 key에 들어갈 값으로는 ("반드시 인자가하나, 정렬에사용가능한 값하나"로 구성된 함수와, 또 다른 콜러블)만 허용하고 나머지는 예외처리를 해놨을 것이다.
그리고 reverse는 True, False 만 받게끔 돼있을 것이다.

이 함수를 실행하게 되면 콜러블이된 "key"의 인자값으로 iterable을 차례로 넣어주면서 값을 비교한 후 정렬해서 리턴을 주는 구조일 것같다..
```
