```python
#################### configparser ####################


import configparser

user_conf = configparser.ConfigParser()

user_conf["DEFAULT"] = {
    "human" : "yes",
    "reference" : "home"
}

user_conf["user_1"] = {
    "name" : "hrvy",
    "gender" : "male",
    "reference" : "school",
    "age" : "10,10.9",
    "job" : "singer"
}


user_conf["user_2"] = {}
user_conf["user_2"]["name"] = "harry"
user_conf["user_2"]["gender"] = "male"
user_conf["user_2"]["age"] = "20"
user_conf["user_2"]["job"] = "singer"


user_conf["user_3"] = {}
user_3 = user_conf["user_3"]
user_3["human"] = "no"
user_3["name"] = "moongchi"
user_3["gender"] = "female"
user_3["age"] = "2.9"
user_3["job"] = "dog"
#configparser 모듈을 써서 config파일을 만들어주는 양식이다. 방법은 총 3가지가 있고, user_1,2,3 순으로 3가지 방법을 다써봤다.
#맨위에있는 DEFAULT 섹션은 모든 섹션에 초기화값으로 설정된다.(섹션별 따로 지정값이 없을 경우)
#모듈의 사용방식이 dict와 흡사하다.(같진않음)
#user_1은 age의 값으로 리스트를 넣어줬다.

with open("md_configparser_conf.conf", mode="w", encoding="utf8")as file:
    user_conf.write(file)
user_conf.write(open("user_conf.conf","w",encoding="utf8")) 이렇게 간단하게 할 수도 있다.
#open함수를 이용해서 "user_conf"인스턴스를 conf파일로 저장을 하는데,
# "인스턴스명.write(as file)" 이런식으로 동작하는것을 확인할 수 있다.
#확장자명는 ".ini", ".conf(=.cfg)", ".txt" 이렇게 사용해도 전부 무방하다.
#txt는 직접 써보지않아서 확실하진않음.(사실 config(initialize)파일임을 표현하기위해서라도 txt는 안쓰는게 맞다고본다.)


################################### ⬇️ md_configparser_conf.conf ⬇️  ##########################################
[DEFAULT]
human = yes
reference = home

[user_1]
name = hrvy
gender = male
reference = school
age = 10,10.9
job = singer

[user_2]
name = harry
gender = male
age = 20
job = singer

[user_3]
human = no
name = moongchi
gender = female
age = 2.9
job = dog

################################### ⬆️ md_configparser_conf.conf ⬆️  ##########################################
```

```python

conf_r = configparser.ConfigParser()
print(conf_r.read("md_configparser_conf.conf"))
print(conf_r.sections())
print ("user_1" in conf_r)
print("user_4" in conf_r)
################################### ⬇️SHELL⬇️ ##########################################
"""
<<SHELL>> python md_configparser.py

['md_configparser_conf.conf']
['user_1', 'user_2', 'user_3']
True
False
"""
################################### ⬆️SHELL⬆️ ##########################################
#config파일을 생성할때와는 다르게 불러올 떄는 인스턴스를만들어준뒤, "인스턴스명.read("파일명(당연히 경로도 상관없다.)")"만으로 불러올 수 있다.
#read메서드로 파일을 불러오게되면 인스턴스에 파일의 주소값이 저장된다.
#불러온 파일명, 섹션들은 list로 불러온다.



conf_r = configparser.ConfigParser()
path = "/Users/jaeha/Desktop/study/EX_codes/md_configparser_conf.conf"
conf_r.read(path)
print(list(conf_r["user_3"].items()))
print(dict(conf_r["user_3"].items()))
user_2 = conf_r["user_2"]
print(user_2["name"])
print(conf_r["user_3"]["human"])
print(conf_r["user_3"]["reference"])
print(conf_r["user_3"]["job"])
################################### ⬇️SHELL⬇️ ##########################################
"""
<<SHELL>> python md_configparser.py
[('human', 'no'), ('name', 'moongchi'), ('gender', 'female'), ('age', '2'), ('job', 'dog'), ('reference', 'home')]
{'human': 'no', 'name': 'moongchi', 'gender': 'female', 'age': '2.9', 'job': 'dog', 'reference': 'home'}
harry
no
home
dog
"""
################################### ⬆️SHELL⬆️ ##########################################
#.read()메서드에 경로를 입력했다.
#config파일은 섹션별로 dict형태의 내용을 가지고있기때문에, list,tuple,dict 로 값을 바로 가져올 수 있다.
#각 섹션을 객체화하는것도 가능하다.
#user_3에대한 값을보면 human은 default값이아닌 직접정한값이 나왔고, reference는 따로 지정하지않았기때문에 디폴트값으로 출력됐다.



conf_r = configparser.ConfigParser()
path = "/Users/jaeha/Desktop/study/EX_codes/md_configparser_conf.conf"
conf_r.read(path)

print([{k : v} for k ,v  in conf_r["user_1"].items()])
print(conf_r["user_1"]["age"])
print(conf_r["user_1"]["age"].split(",")[0])
print(conf_r["user_1"]["age"].split(",")[1])
print(int(conf_r["user_2"]["age"]))
print(float(conf_r["user_3"]["age"]))
################################### ⬇️SHELL⬇️ ##########################################
"""
<<SHELL>> python md_configparser.py

[{'name': 'hrvy'}, {'gender': 'male'}, {'reference': 'school'}, {'age': '10,10.9'}, {'job': 'singer'}, {'human': 'yes'}]
10,10.9
10
10.9
20
2.9
"""
################################### ⬆️SHELL⬆️ ##########################################
#user_1섹션의 값을 items()로 받아온 뒤 list conprehention(list expression)으로 만들어봤다.(dict형식이라 가능하다.)
#user_1의 나이입력 시 "10,10.9" 이렇게 입력했는데, configparser로 파일을 만들때 값들은 전부 str로 처리해야하기 때문에
#이런방법으로 입력해준뒤, 불러올 때는 .split(",")로 ,기준으로 나눈 리스트로 만들어준 뒤 인덱싱한다.(좋은방법인지는 모르겠다...)
#각자료형의 형식에 부합한다면, int,float으로 변환해서 사용할 수 있다.(강제변환이된다.)


conf_r = configparser.ConfigParser()
path = "/Users/jaeha/Desktop/study/EX_codes/md_configparser_conf.conf"
conf_r.read(path)

user_1 = conf_r["user_1"]
user_2 = conf_r["user_2"]
print(user_1.getboolean("human"))
print(user_2.getint("age"))
print(conf_r.getboolean("user_3","human"))
print(conf_r.getfloat("user_3","age"))
################################### ⬇️SHELL⬇️ ##########################################
"""
<<SHELL>> python md_configparser.py

True
20
False
2.9
"""
################################### ⬆️SHELL⬆️ ##########################################
#user_1은 디폴트값으로 hummen = yes를 가지고 있다.
#getint, getfloat역시 사용가능하다.
#.getboolean()메서드는 'yes'/'no', 'on'/'off', 'true'/'false', '1'/'0'를 대소문자 구분없이 True/False로 구분한다!


conf_r = configparser.ConfigParser()
path = "/Users/jaeha/Desktop/study/EX_codes/md_configparser_conf.conf"
conf_r.read(path)

user_1 = conf_r["user_1"]
print(user_1.get("job"))
print(user_1.get("job", "teacher"))
print(user_1.get("like", "food"))
print(user_1.get("dream", fallback="president"))
################################### ⬇️SHELL⬇️ ##########################################
"""
<<SHELL>> python md_configparser.py

singer
singer
food
president
"""
################################### ⬆️SHELL⬆️ ##########################################
#get()메서드를 사용하면 키에 맞는 값을 얻을 수 있는데, 이때 fallback(대체값)을 입력을 바로해줄 수 있다.
#물론 키에 해당하는 값이 존재하거나, 디폴트로 가져올 수 있다면 fallback은 적용되지 않고,
#값이 없을 시 적용이된다, fallback키워드를 사용하면 된다.
```
