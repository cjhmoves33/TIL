```python
#################### Data Structure ####################

#Stack & Queue

#Stack구조는 후입선출(LIFO - Last In First Out)방식이다.data 입력을 push(append) 출력을 pop(pop)이라고 부른다.
#파이썬의 List자료형이 stack구조와 같다.

a=[1,2,3,4]
a.append(5)
>> [1,2,3,4,5]
a.pop()
>> 5
#실제론 append메소드는 리턴값이 없기때문에 출력값이없다. [1,2,3,4,5]은 가독성을위해서 적은 것.
#pop메소드는 return값을 가진다. 

#Queue구조는 선입선출(FIFO- First In First Out)방식이다. Stack과 반대.
a=[1,2,3,4,5]
a.append(6)
frst = a.pop(0)
a
[2,3,4,5,6]
frst
6
#pop(0)으로 큐방식으로 값을 추출할 수있다. 리스트의 0번째 값을 pop한것.



#Set
#set(집합)자료형은 데이터 순서와 성관없이 저장하고, 중복을 불허한다.
a={1,2,3}
b= set([4,5,6])
a.add(1)
print(a)
print(b)
type(a)
>> {1,2,3}
   {4,5,6}
    set
#a의 값을보면 1을 추가로 더해줬는데 값이변하지않았다. 중복을 불허한다.
b = set([5,6,7])
b.add(1)
b.update([2,3])
b.remove(1)
b.discard(2)
print(b)
b.clear()
print(b)
>> {3,5,6,7}
   set()
#add는 값하나를 추가,update로 여러개의 값을 추가
#remove는 인자값이 집합안에 없으면 에러, discard는 인자값이 집합안에 없어도 에러가없음.
#clear는 빈 집합으로 만들어준다.
set_1 = {1,2,3,4,5}
set_2 = {3,4,5,6,7}

union = set_1.union((set_2))
union_same = set_1|set_2
>> {1,2,3,4,5,6,7}
#합집합

intersection = set_1.intersection(set_2)
intersection_same = set_1&set_2
>> {3,4,5}
#교집합

diff = set_1.difference(set_2)
diff_same = set_1 - set_2
>> {1,2}
#차집합
#실제 집합연산이 가능하다.
```
