```python
#################### logging ####################


import logging
logging.warning("warning!")
logging.info("information!")
#logging모듈을 임포트해서 실행시킨다.
#logLevel은 기본적으로 debug->info->warning->error->critical 다섯가지가 있고, 각 단계는 중요도 순으로 나열된 것


################################### ⬇️SHELL⬇️  ##########################################
"""
python lgg.py (현위치에 lgg.py파일로 존재한다고 가정, 입력 시)
WARNING:root:warning!
"""
################################### ⬆️SHELL⬆️  ##########################################
#logging모듈의 기본레벨은 warning 이다. 달리 레벨을 수정하지않을 시 이 단계 이상만 추적(사용)가능.(warning->error->critical)



import logging
logging.basicConfig(filename="ex.log", encoding="utf8", level=logging.DEBUG)
logging.debug("디버그메세지")
logging.info("인포메세지")
logging.warning("워닝메세지")
logging.error("에러메세지")
logging.critical("크리티컬메세지")
#현위치에 ex.log라는 파일이 생성됐고, 모든 레벨의 메세지가 입력됐다.
#basicConfig에도 파일생성, 포매팅 등등 입력할 수있으나, 어차피 각 로깅인스턴스마다 핸들러를 추가할 것이기 때문에,
#실질적으로는 level값만 넣어준다. 3.8이상부터는 basicConfig로 기본레벨을 먼저 설정해줘야하기 때문.
#물론 스트림핸들러가 추가되면 console상에선 필요가 없어질 수도 있다. basicConfig은 일회용으로 쓰인다는 말이있다.



import logging
warn = "warning"
msg = "message"
flt = 3.1415925

logging.basicConfig(level=logging.DEBUG)
lgg = logging.getLogger("tester")
lgg.setLevel(level=logging.INFO)
lgg.debug("디버그 메세지")
lgg.info("%10s %s - 인포메세지","info","message")
lgg.warning(f"{warn} {msg} - 워닝메세지")
lgg.error("{0} {1} - 에러 메세지".format("error","message"))
lgg.critical(f"{flt:.2f} - 크리티컬메세지")
################################### ⬇️SHELL⬇️  ##########################################
"""
python lgg.py (입력 시)
INFO:tester:      info message - 인포메세지
WARNING:tester:warning message - 워닝메세지
ERROR:tester:error message - 에러 메세지
CRITICAL:tester:3.14 - 크리티컬메세지
"""
################################### ⬆️SHELL⬆️  ##########################################
#lgg로거를 생성한후 레벨을 다시 설정해줬기때문에(INFO) 디버그레벨은 추적할 수 없다.
#출력값이 전부 str이기때문에 각종 문자열 포멧이 가능하다.
#lgg로거를 생성할때 로거명을 "tester"로 지정해줬기때문에 이제 root가 아닌 로거명으로 출력된다.(lgg는 객체일뿐.(사실 인스턴스가 정확한표현))


#lgg.py
import logging
flt = 3.1415925
logging.basicConfig(level=logging.DEBUG)
def get_lgg(nm):
    lgg = logging.getLogger(nm)
    lgg.setLevel(level=logging.INFO)
    lgg.critical(f"{flt:.2f} - 크리티컬메세지")
#tst1.py
import lgg
lgg.get_lgg(__name__)
#tst2.py
import tst1

################################### ⬇️SHELL⬇️  ##########################################
"""
python tst2.py (입력 시)
CRITICAL:tst1:3.14 - 크리티컬메세지
"""
################################### ⬆️SHELL⬆️  ##########################################
#lgg.py에서 로거를 만드는 함수를 만들고, tst1.py에서 모듈을임포트해서 로그명이(__name__)인 로거를만들었다.
#이제 tst2.py에서 tst1.py를 임포트해보면 로거명이tst1으로 된걸 확인할 수 있다.
#이런식으로 __name__을 활용해서 로그의 발생경로를 알 수도 있다.

import logging
flt = 3.1415925
FORMAT = "%(asctime)s  %(name)s %(levelname)s"
FORMAT_dt = "%Y/%m/%d %H:%M:%S"

logging.basicConfig(level=logging.DEBUG)
formatter = logging.Formatter(fmt=FORMAT, datefmt=FORMAT_dt)
hdl = logging.StreamHandler()
hdl.setFormatter(formatter)

lgg = logging.getLogger("main")
lgg.setLevel(level=logging.INFO)
lgg.addHandler(hdl)

lgg.critical(f"{flt:.2f} - 크리티컬메세지")

lgg2 = logging.getLogger("main.fst")
lgg.warning("test")
#포맷할 내용을 미리 객체화시킨 후, 포매터에 설정해준다.
#로거를 생성할때 이미존재하는 로거명뒤에 . 찍게되면 그 로거의 자식로거를 생성할 수 있다.
#핸들러,포맷등 적용된것들이 전부 넘어온다.


import logging

flt = 3.1415925
FORMAT = "%(asctime)s  %(name)s %(levelname)s %(message)s"
FORMAT_dt = "%Y/%m/%d %H:%M:%S"

logging.basicConfig(level=logging.DEBUG)

hdl_s = logging.StreamHandler()
hdl_f = logging.FileHandler(filename="ex_logging.log",mode="w", encoding="utf8")

hdl_s.setFormatter(formatter)
hdl_s.setLevel(level=logging.INFO)
hdl_f.setFormatter(formatter)
hdl_f.setLevel(level=logging.ERROR)

lgg = logging.getLogger("main")
lgg.addHandler(hdl_s)
lgg.addHandler(hdl_f)

lgg.critical(f"{flt:.2f} - 크리티컬메세지")

lgg2 = logging.getLogger("main.fst")
lgg2.warning("test")
#filehandler까지 추가해줬다.
#이제 streamhandler가 있기 때문에 basicConfig는 없애준다.(지금으로선 필요가없다.)
#각 handler(== 핸들러, 처리기)별로 레벨을 설정 해줘도된다.
#여느 파일핸들러들과 같이 filename에는 경로가 들어가도 된다.
#ex_logging.log파일을 만들고 아래 내용이 입력됐다.
#2021/06/25 15:34:44  main CRITICAL  3.14 - 크리티컬메세지
#2021/06/25 15:34:44  main.fst WARNING  test

```

```python


#위에서 했던 방법처럼 로그를 파이썬에서 만들어 줄 수 있고,
#JSON파일로 만들어놓을수도 있다.

{
    "version": 1,
    "disable_existing_loggers": false, 
    "formatters": {
        "s_fmt": {
            "format": "%(name)s  %(levelname)s\n%(message)s",
            "datefmt": "%Y/%m/%d - %H:%M:%S"
        },
        "f_fmt": {
            "format": "%(asctime)s  %(name)s  %(levelname)s\n%(message)s",
            "datefmt": "%Y/%m/%d - %H:%M:%S"
        }
    },
    "handlers": {
        "s_hdl": {
            "class": "logging.StreamHandler",
            "level": "DEBUG",
            "formatter": "s_fmt"
        },
        "f_hdl": {
            "class": "logging.FileHandler",
            "level": "DEBUG",
            "formatter": "f_fmt",
            "filename": "file.log",
            "mode": "a"
        }
    },
    "loggers": {
        "__main__": {
            "level": "DEBUG",
            "handlers": ["s_hdl","f_hdl"],
            "propagate": true
        }
    }
}
#version은 1을 주면 된다.
#disable_existing_loggers는 fileConfige나 dictConfig를 활성화시켰을 때, 기존의 로거들이 전부비활성화된다.(일단 false로 해두자.)
#단순히 streamg핸들러와 file핸들러로 구분해서 하나씩 만들어봤다. 이제 파이썬에서 불러와서 적용시켜보자.


import logging
import logging.config
import json

confg = json.load(open("loggin_ex.json"))
logging.config.dictConfig(confg)
logger = logging.getLogger(__name__)

logger.info("info -- test--")
#json파일을 불러서 객체화시켜준뒤, dictConfig의 인자로 넣어줬다.
#불러온파일로 로거를 활성화시켜주기위해서는 json안에있는 loggers영역의 로거명과 내가만드는 로거명이 일치해야한다.
#어쨋든 __name__ == "__main__" 이니까 여기선 문제없이 작동한다. 좋은이름인지는 모르겠다.

################################### ⬇️file.log⬇️  ##########################################
"""
2021/06/25 - 17:46:06  __main__  INFO
info -- test--
2021/06/25 - 17:46:07  __main__  INFO
info -- test--
"""
################################### ⬆️file.log⬆️  ##########################################

################################### ⬇️SHELL⬇️  ##########################################
"""
python lgg.py
__main__  INFO
info -- test--
python lgg.py
__main__  INFO
info -- test--
"""
################################### ⬆️SHELL⬆️  ##########################################


#외에도 dictConfig의 다양한 활용이있다. 구글링하거나
#https://docs.python.org/ko/3/library/logging.config.html#module-logging.config 참조
#https://hwangheek.github.io/2019/python-logging/ 참조
```
