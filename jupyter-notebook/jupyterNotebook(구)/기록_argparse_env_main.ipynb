```python
#################### argparse(command - line option) ####################
#이 모듈을 이용하면 .py파일을 인터프리터에 돌렸을 때 command line option을 사용할 수 있다.

import argparse
#argparse모듈을 임포트한다.

parser = argparse.ArgumentParser() 
#camel case는 class. "parser"인스턴스를 만들어줌으로서 command line option을 사용할 수 있게 됐다.
#아직은 아무것도 나오지 않는다.
args = parser.parse_args()
#!!!!!!!!!!      실제 환경에서 실행시킬 때는 이 위치에 절대 있으면 안된다       !!!!!!!!!!

#parse_args()메서드를 실행시키면 "앞으로 사용하게될" 옵션(command line option)들의 값을 받아서 리턴시켜준다!(여기있으면 안되는 이유)
#사실 이 기능만 활성히켜도 기본적으로 -h 옵션이 생기는데 shell에서 실행시켜보면
#(지금부터 바탕화면에 apr.py로 파이썬파일이 존재한다는 가정하에)

```

```python


#python arp.py -h 입력 시

#usage: arp.py [-h] 사용할 수 있는 옵션들이 나온다. 기본적으로 -h가 있다.
#optional arguments:
#  -h, --help  show this help message and exit

#이제 옵션을 만들어보자.
parser.add_argument("echo", help = "just test!")
#<<SHELL>> python arp.py

#usage: arp.py [-h] echo "echo"가 추가된게 보인다.
#arp.py: error: the following arguments are required: echo 이 부분은 다음파트에서 설명한다.

#python arp.py -h 이제 help를 입력하게 되면

################################### ⬇️SHELL⬇️  ##########################################
"""
<<SHELL>> python arp.py -h

usage: arp.py [-h] echo

positional arguments:
 echo        just test!

optional arguments:
 -h, --help  show this help message and exit
 """
################################### ⬆️SHELL⬆️  ##########################################


#positional과 optional을 구분하는 방법은 옵션 생성때 사용할 옵션이름앞에 -를 붙혀주면 된다.(-은 짧은이름 --은 긴이름,, 둘다 써도 된다.)
#바로 적용해보겠다.

parser.add_argument("-echo2", "--echo2", help = "just test!")


################################### ⬇️SHELL⬇️  ##########################################
"""
usage: arp.py [-h] [-echo2 ECHO2] echo 

positional arguments:
  echo                  just test!

optional arguments:
  -h, --help            show this help message and exit
  -echo2 ECHO2, --echo2 ECHO2
                        just test!
"""
################################### ⬆️SHELL⬆️  ##########################################
#-echo2는 "optional"로 추가했기때문에 이름이 []로 묶여버렸다.
#여기까지 간단히 알아 보고 기능을 수행할 수 있는 옵션들을 만들어 보자.

```

```python

#전 파트에서  arp.py: error: the following arguments are required: echo  가 있었는데,
#positional옵션은 인터프리터로 파일을 열때 반드시 파일명 뒤에 인자값을 넣어줘야한다.
#optional옵션은 선택가능(required = True/False 기본값은 False)
#이 내용을 적용해서 옵션들을 만들것이다.

parser.add_argument("square", help="제곱값을 돌려준다.", type=int)
parser.add_argument("plus_1", help="입력값 더하기 1을 돌려준다..", type=int)
#여기서 타입을 설정해주는 이유는 argparse는 우리가 만든 옵션들을 str로 취급하기 때문이다.
parser.add_argument("-mi1","--minus_1", help="입력값 빼기 1을 돌려준다..", type=int, dest="mn_1")
parser.add_argument("-mi2", "--minus_2", help="입력값 빼기 2을 돌려준다..", type=int)
#dest는 parse_args()가 값을 반환할 때 그 값이 저장될 객체이름.

args = parser.parse_args()
#설명은 위에서 했지만, 인터프리터 특성상 옵션들을 활성시키고 값을 받기위해선 이 메서드가 반드시 add_argument()작업을 끝낸 후 그뒤에 위치해야한다.

print(args) #namespace형식으로 출력된다.
print(args.square**2)
print(args.plus_1+1)
print(args.mn_1-1) #dest로 지정한 객체이름
print(args.minus_2-2)#dest를 해주지않으면 긴이름을 기준으로 객체가 생성된다.(없으면 짧은 이름) 

################################### ⬇️SHELL⬇️  ##########################################
"""
<<SHELL>> python arp.py -mi2=2 -mi1=2 2 2

Namespace(minus_1=2, minus_2=2, plus_1=2, square=2)
4
3
1
0
"""
################################### ⬆️SHELL⬆️  ##########################################
#optional값입력시 붙혀쓴다, -min2 2 -min1 2 2 2 이렇게 입력해도 무관하다.
#optional과 positional을 입력하는 순서는 상관없다, positional은 옵션을 만들어준 순서대로 입력한다.(이것만 잘 지켜주면 되는듯하다.)
```

```python


#이제 optional을 좀더 살펴보자.

import argparse
parser = argparse.ArgumentParser()
parser.add_argument("--verbosity", help="increase output verbosity")
args = parser.parse_args()

if args.verbosity:
    print("verbosity turned on")

# --을 이용해서 optional로 만들어줬다.

################################### ⬇️SHELL⬇️  ##########################################
"""
<<SHELL>> python arp.py
(공백)
<<SHELL>> python arp.py --help
usage: arp.py [-h] [--verbosity VERBOSITY]

optional arguments:
  -h, --help            show this help message and exit
  --verbosity VERBOSITY
                        increase output verbosity
("optional"에 추가된것이 보인다.)


python3 prog.py --verbosity 1
verbosity turned on


<<SHELL>> python arp.py --verbosity
usage: arp.py [-h] [--verbosity VERBOSITY]
arp.py: error: argument --verbosity: expected one argument
"""
################################### ⬆️SHELL⬆️  ##########################################

#위의 코드를 보면 --verbosity의 값이 있으면 무언가를 프린트를 하게끔 돼있다.(아무값이나 넣어줘도 무관하다. 현재if문의 조건상 값의 존재여부만 봄.)
#이 파일을 쉘에서 실행시 아무것도 입력하지 않으면 오류가없다.(optional로 추가했기때문)
#그리고 아무것도 입력하지 않았기 때문에 --verbosity의 관련변수의 값에는(이 경우엔 args.verbosity(--verbosity의값이 저장돼어야할 객체))
#None이 주어지게된다. 때문에 if문은 false가 돼고 아무것도 출력되지 않는것이다.
#하지만 맨 밑 쉘부분을 보면 에러가 나온것을 확인할 수 있다. 쉘에서 --verbosity를 실행하기위해 입력 했다면 반드시 값을 넣어줘야한다.(기본설정)
#물론 위의 코드의 경우 --verbosity는 true or false 만이 필요로 하기에 무언가 추가를 해줘야한다.




import argparse
parser = argparse.ArgumentParser()
parser.add_argument("--verbosity", help="increase output verbosity", action="store_true")
args = parser.parse_args()

if args.verbose:
    print("verbosity turned on")
# --verbosity 를 만들어줄 때 action="store_true"를 추가해줬다.

################################### ⬇️SHELL⬇️  ##########################################
"""
<<SHELL>> python3 arp.py --verbose
verbosity turned on

<<SHELL>> python3 arp.py --verbose 1
usage: arp.py [-h] [--verbose]
arp.py: error: unrecognized arguments: 1

<<SHELL>> python arp.py --help
usage: arp.py [-h] [--verbosity]

optional arguments:
  -h, --help   show this help message and exit
  --verbosity  increase output verbosity
"""
################################### ⬆️SHELL⬆️  ##########################################
#--verbosity를 만들어줄때 새로운키워드인 action을 추가하고 값응로 "store_true"를 넣어줬다.
#이렇게 되면 옵션을 사용하기위해 쉘에 입력할때 args.verbosity(이 옵션의경우)에 자동으로 True의값을 넣어주게된다.
#설정이 변경된 것이다. 이젠 특정값과 함께 입력시 에러가 뜬다.
#-h/--help에서 보이는 것도 바뀐게 보인다. 이제 옵션명뒤에 대문자이름이 사라졌다. 값이 무조건 True로 작용하기때문에 없애준것같다.



import argparse
parser = argparse.ArgumentParser()
parser.add_argument("-v", "--verbosity", help="increase output verbosity", action="store_true")
parser.add_argument("square", help="제곱값을 나타냄", type=int)

args = parser.parse_args()
answer = args.square**2

if args.verbosity:
    print(f"{args.square}의 제곱은 {answer} ")
else:
    print(answer)
    
################################### ⬇️SHELL⬇️  ##########################################
"""
<<SHELL>> python arp.py 2
4

<<SHELL>> ython arp.py 2 -v
2의 제곱은 4
"""
################################### ⬆️SHELL⬆️  ##########################################
#이런 식으로 옵션들을 혼합해서 사용할 수도 있다.




import argparse
parser = argparse.ArgumentParser()
parser.add_argument(
    "-v", "--verbosity", help="increase output verbosity",
     type=int, choices=[0, 1, 2]
     )
parser.add_argument("square", help="제곱값을 나타냄", type=int)

args = parser.parse_args()
answer = args.square**2

if args.verbosity == 2:
    print(f"{args.square}의 제곱은 {answer} ")
elif args.verbosity ==1:
    print(f"{args.square}^2 == {answer}")
else:
    print(answer)
    
#이번에는 choises키워드를 추가해줬다.
#이제 -v를 사용하려면 값을 반드시 입력해줘야하고 그 값으로 0, 1, 2 중에서만 선택 할 수 있게된다.
#쉘에서 입력하면 그에 맞는 결과들이 프린트될것이다.


import argparse
parser = argparse.ArgumentParser()
parser.add_argument(
    "-v", "--verbosity", help="increase output verbosity",
    type=int
    )
parser.add_argument("square", help="제곱값을 나타냄", type=int)

args = parser.parse_args()
answer = args.square**2

if args.verbosity >= 2:
    print(f"{args.square}의 제곱은 {answer} ")
elif args.verbosity ==1:
    print(f"{args.square}^2 == {answer}")
else:
    print(answer)
    
################################### ⬇️SHELL⬇️  ##########################################
"""
<<SHELL>> python arp.py 2
Traceback (most recent call last):
  File "/Users/jaeha/Desktop/arp.py", line 12, in <module>
    if args.verbosity >= 2:
TypeError: '>=' not supported between instances of 'NoneType' and 'int'
"""
################################### ⬆️SHELL⬆️  ##########################################
#문제가 생겼다. 옵션을 사용하지않을경우 값이None으로 지정되는데, if문에서 비교연산자를 만나게 되면서 오류가 발생한 것이다.
#-v의 값이 2 이상인지 비교를 해야하는데 None 과 int는 비교를 할 수가없다.




import argparse
parser = argparse.ArgumentParser()
parser.add_argument(
    "-v", "--verbosity", help="increase output verbosity",
    type=int, default=0
    )
parser.add_argument("square", help="제곱값을 나타냄", type=int)

args = parser.parse_args()
answer = args.square**2

if args.verbosity >= 2:
    print(f"{args.square}의 제곱은 {answer} ")
elif args.verbosity ==1:
    print(f"{args.square}^2 == {answer}")
else:
    print(answer)
#default키워드를 추가해줬다. 옵션을 사용하지않았을 때 기본값이 0 이되는 마법.
#이제 -v를 입력해주지 않아도 기본값이 int(이 경우엔 0)이기때문에 if문릉ㄹ 만나도 문제가 없다.
################################### ⬇️SHELL⬇️  ##########################################
"""
<<SHELL>> python arp.py 2
4
"""
################################### ⬆️SHELL⬆️  ##########################################

#argparse에대한 간단한 내용은 여기까지다. 자세한 내용은
#https://docs.python.org/ko/3.9/howto/argparse.html?highlight=argparser
#https://docs.python.org/ko/3.9/library/argparse.html#module-argparse
```
